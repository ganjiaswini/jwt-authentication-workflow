```python
import jwt
import datetime

# Secret Key
SECRET_KEY = "my_secret_key"

print("=" * 50)
print("JWT AUTHENTICATION WORKFLOW DEMO")
print("=" * 50)

# Generate Access Token
def generate_access_token(user_id):
    payload = {
        "user_id": user_id,
        "token_type": "access",
        "exp": datetime.datetime.utcnow() + datetime.timedelta(minutes=15)
    }

    return jwt.encode(payload, SECRET_KEY, algorithm="HS256")


# Generate Refresh Token
def generate_refresh_token(user_id):
    payload = {
        "user_id": user_id,
        "token_type": "refresh",
        "exp": datetime.datetime.utcnow() + datetime.timedelta(days=7)
    }

    return jwt.encode(payload, SECRET_KEY, algorithm="HS256")


# Validate Token
def validate_token(token):
    try:
        payload = jwt.decode(
            token,
            SECRET_KEY,
            algorithms=["HS256"]
        )

        print("\n Token Valid")
        print("Payload:", payload)

        return payload

    except jwt.ExpiredSignatureError:
        print("\n Token Expired")

    except jwt.InvalidTokenError:
        print("\n Invalid Token")


# Generate New Access Token using Refresh Token
def refresh_access_token(refresh_token):
    try:
        payload = jwt.decode(
            refresh_token,
            SECRET_KEY,
            algorithms=["HS256"]
        )

        if payload["token_type"] != "refresh":
            print("Invalid Refresh Token")
            return None

        new_access_token = generate_access_token(
            payload["user_id"]
        )

        print("\n New Access Token Generated")
        return new_access_token

    except Exception as e:
        print("Error:", e)


# -----------------------------
# Authentication Workflow Demo
# -----------------------------

print("\nSTEP 1: User Login")
user_id = 101

access_token = generate_access_token(user_id)
refresh_token = generate_refresh_token(user_id)

print("\nAccess Token:")
print(access_token)

print("\nRefresh Token:")
print(refresh_token)

print("\n" + "=" * 50)
print("STEP 2: Access Protected Resource")
print("=" * 50)

validate_token(access_token)

print("\n" + "=" * 50)
print("STEP 3: Refresh Access Token")
print("=" * 50)

new_access_token = refresh_access_token(refresh_token)

print("\nNew Access Token:")
print(new_access_token)

print("\n" + "=" * 50)
print("STEP 4: Validate New Access Token")
print("=" * 50)

validate_token(new_access_token)

print("\n" + "=" * 50)
print("JWT AUTHENTICATION WORKFLOW COMPLETED")
print("=" * 50)
```
